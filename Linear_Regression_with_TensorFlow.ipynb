```python
# Installing the required Libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data: y = 3.5x + 2 + noise
np.random.seed(42)
X = np.linspace(0, 10, 100)
Y = 3.5 * X + 2 + np.random.randn(*X.shape) * 2  # add some noise

# Plot
plt.scatter(X, Y)
plt.title("Generated Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Create a simple linear regression model using tf.keras
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Compile model with Mean Squared Error loss and SGD optimizer
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
              loss='mean_squared_error')

# Train the model (fit)
history = model.fit(X, Y, epochs=200, verbose=0)

# Plot training loss
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# Print learned weights and bias
weights, bias = model.layers[0].get_weights()
print(f"Learned Weight: {weights[0][0]:.4f}, Bias: {bias[0]:.4f}")

# Predict using the trained model
Y_pred = model.predict(X)

# Plot actual vs predicted
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.legend()
plt.title('Linear Regression Fit')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
